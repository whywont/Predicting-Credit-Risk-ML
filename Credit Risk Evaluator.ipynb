```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
train_df
```

```python
test_df
```

```python
# Convert categorical data to numeric and separate target feature for training data
X = train_df.drop('loan_status', axis=1)
train_dummies = pd.get_dummies(X)

label = LabelEncoder().fit_transform(train_df['loan_status'])

print(label)
```

```python
train_dummies
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_test = test_df.drop('loan_status', axis=1)
test_dummies = pd.get_dummies(X_test)

label_test = LabelEncoder().fit_transform(test_df['loan_status'])
print(label_test)
```

```python
test_dummies
```

```python
print(label_test)
```

```python
# Get missing columns in the training test
missing_cols = set( train_dummies.columns ) - set( test_dummies.columns )
# Add a missing column in test set with default value equal to 0
for c in missing_cols:
    test_dummies[c] = 0
test_dummies = test_dummies[train_dummies.columns]
```

```python
test_dummies
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(train_dummies, label)
```

```python
print(f"Training Data Score: {classifier.score(train_dummies, label)}")
print(f"Testing Data Score: {classifier.score(test_dummies, label_test)}")
```

# Trying with train_test_split and string column values

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1, n_estimators=300).fit(train_dummies, label)
print(f'Training Score: {clf.score(train_dummies, label)}')
print(f'Testing Score: {clf.score(test_dummies, label_test)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(train_dummies)
train_dummies_scaled = scaler.transform(train_dummies)

scaler_test = StandardScaler().fit(test_dummies)
test_dummies_scaled = scaler_test.transform(test_dummies)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier_scaled = LogisticRegression()
classifier_scaled.fit(train_dummies_scaled, label)

print(f"Training Data Score: {classifier_scaled.score(train_dummies_scaled, label)}")
print(f"Testing Data Score: {classifier_scaled.score(test_dummies_scaled, label_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf_scaled = RandomForestClassifier(random_state=1, n_estimators=300).fit(train_dummies_scaled, label)
print(f'Training Score: {clf_scaled.score(train_dummies_scaled, label)}')
print(f'Testing Score: {clf_scaled.score(test_dummies_scaled, label_test)}')
```

```python
predictions = classifier_scaled.predict(train_dummies_scaled)
pd.DataFrame({"Prediction": predictions, "Actual": label})
```
